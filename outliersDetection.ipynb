# Outliers 

Outliers are observations that are significantly distant from other observations.

Outliers are points that deviate from the general trend of the data set or diverge significantly from other observations. These outliers negatively affect the orientation of the data set.

It causes deviations in the statistical calculations of the data set, so outliers should be detected before data analysis
and those outliers should be filled with appropriate values.

```python
#include libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#create an example DataFrame
```

```python
df = pd.DataFrame({
    'A': [28, 34, 40, 45, 68, 65, 82, 91, 110, 200]
})
```

```python
#Box plot is a type of chart that allows us to visually observe the quartiles and outliers of the data set. 
#This type of chart is used to summarize the distribution of the data set and identify outliers.
```

```python
#Draw the box plot
```

```python
plt.boxplot(df)
plt.title('Observing Outliers') 
plt.ylabel("Column A")
plt.show()
```

```python
#The box plot graph shows the minimum, first quartile (Q1), median, third quartile (Q3) and maximum values of the data. 
#It also contains lines called 'whiskers' that help identify potential outliers.
```

# 

```python
#Whiskers usually represent the limits Q1=Q1 - 1.5 * IQR and Q3=Q3 + 1.5 * IQR (IQR: Q3 - Q1). Values outside 
#these limits may be considered outliers.
```

```python
#On the box plot, points outside the whiskers boundaries are marked as outliers.
```

```python
# Identify outliers
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = df[(df < lower_limit) | (df > upper_limit)].dropna()
```

```python
print("Outliers:")
print(outliers)
```

Identifying and handling outliers is an important step in the data analysis and modeling processes. This process should be carefully evaluated and an appropriate approach should be determined according to the purposes of the analysis and the nature of the data set.
